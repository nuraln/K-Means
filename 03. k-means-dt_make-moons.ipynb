## DBSCAN Clustering

DBSCAN groups together data points that are close to each othe based on their density, allowing for flexibe cluster shapes and handling varying densities

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Generate data
X, y = make_moons(n_samples=250, noise=0.10, random_state=22)

# display the generate data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Original data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```

### Feature Scaling (Standard Scaler)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbcan = DBSCAN(eps=0.28)
dbcan.fit(X_scaled)

labels = dbcan.labels_
```

```python
dbcan.labels_
```

```python
# Plot the original data points with their true labels
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Original data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```

```python
# Plot the clusters identified by DBSCAN
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Original data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```
